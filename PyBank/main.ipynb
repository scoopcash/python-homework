```python
#importing pandas and pathlib
import pandas as pd
from pathlib import Path
```

```python
#declaring path to csv doc
csvpath = Path("Resources/budget_data.csv")
```

```python
#createing datafrom to hold data from csv
financials_df = pd.read_csv(csvpath)
```

```python
#checking contents of dataframe
financials_df
```

```python
#cleaning the column names
financials_df.columns=('date', 'pnl')
#running command to get details of dataframe
financials_df.describe(include='all')
```

```python
#* The total number of months included in the dataset.
#from statistics import mean
#comands to get data from df. total, balance, changes, average
total = len(financials_df)
balance = financials_df.pnl.sum()
financials_df['difference'] = financials_df['pnl'].diff()
average_change = financials_df['difference'].mean()
average_change = round(average_change, 2)

#I'm fairly certain the increase and decrease should be written with an iloc or loc statment. but I can't figure it out. 
greatest_increase_profits = financials_df['difference'].max()
greatest_decrease_profits = financials_df["difference"].min()

#print statements to display data
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {total}")
print(f"Total: ${balance}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Profits: ${greatest_increase_profits}")
print(f"Greatest Decrease in Profits: ${greatest_decrease_profits}") 
```
